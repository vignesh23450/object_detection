```python
import cv2 as c
import matplotlib.pyplot as mp



file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen='frozen_inference_graph.pb'
label='labels.txt'


model=c.dnn_DetectionModel(frozen,file)

classLabel=[]

with open(label,'rt')as fpt:
    classLabel=fpt.read().rstrip('\n').split('\n')
```

```python
print( classLabel)
print(len(classLabel))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127,5,127.5))
model.setInputSwapRB(True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

img=c.imread('1200px-Cityroadsyd.jpg')
plt.imshow(img)
```

```python
# Specify the input size
##inputSize = (320, 320)  # Adjust the dimensions according to your model's requirements

# Resize the input image/frame to the specified input size
##resizedImg = c.resize(img, inputSize)

# Perform detection using the resized image/frame
##ClassIndex, confidence, bbox = model.detect(resizedImg, confThreshold=1)






ClassIndex,confidece,bbox= model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale=3
font=c.FONT_HERSHEY_PLAIN
for ClassInd,conf,boxes in zip(ClassIndex.flatten(),confidece.flatten(),bbox):
    c.rectangle(img,boxes,(255,0,0),2)
    c.putText(img,classLabel[ClassInd-1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=3)
```

```python
import cv2 as cv

cap = cv.VideoCapture('E:/Downloads/traffic.mp4')
if not cap.isOpened():
    cap = cv.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open video file.')

font_scale = 3
font = cv.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    ClassIndex, confidece, bbox = model.detect(frame, confThreshold=0.5)
    print(ClassIndex)

    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if ClassInd <= 80:
                cv.rectangle(frame, boxes, (255, 0, 0), 3)
                cv.putText(frame, classLabel[ClassInd-1], (boxes[0]+10, boxes[1]+40),
                           font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

    cv.imshow('objdetection by simplilearn', frame)

    if cv.waitKey(25) & 0xFF == ord('p'):
        break

cap.release()
cv.destroyAllWindows()
```
